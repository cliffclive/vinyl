# Imports and Functions

```python
import os
import pickle
import pandas as pd
import requests

from time import sleep
from random import random
```

```python
def download_genre_mp3_samples(genre_names, genre_metadata_dir, sample_mp3_dir, most=1000):
    for genre in genre_names:
        metadata_path = os.path.join(genre_metadata_dir, f"{genre}_metadata.tsv")
        genre_metadata = pd.read_csv(metadata_path, sep='\t')
        song_previews = dict(zip(genre_metadata.id, genre_metadata.preview_mp3))
        nans = 0
        count = 0
        for k,v in song_previews.items():
            if str(v) == 'nan':
                nans += 1
                continue
            song_id = k
            mp3_url = v
            mp3_filepath = os.path.join(sample_mp3_dir, song_id + '.mp3')
            if os.path.isfile(mp3_filepath):
                continue
            preview = requests.get(mp3_url, allow_redirects=True)
            open(mp3_filepath, 'wb').write(preview.content)
            count += 1
            if count > most: 
                break
            sleep(2*random())
        print(f"{genre} is missing {nans} previews out of {len(song_previews)} songs.")
            
```
